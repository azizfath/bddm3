#implementasi 2

```python
path2 = '/content/drive/MyDrive/Colab Notebooks/DATA MINING 3/week9/bank-full.csv'

df2 = pd.read_csv(path2)
df2.head()
```

```python
print('shape :', df2.shape)
df2.info()
```

```python
#feature selection
cat = []
for col,val in df2.iteritems():
  if val.dtype =='object':
    cat.append(col)

df3 = df2[cat]
```

```python
df3.head()
```

```python
#convert to numeric
from sklearn import preprocessing
le = preprocessing.LabelEncoder()

df_feature_select = df3.copy()
for colName in df_feature_select:
  df_feature_select[colName]=le.fit_transform(df_feature_select[colName])

df_feature_select.head()
```

```python
corr = df_feature_select.corr()

corr_target = abs(corr['housing'])

relevant_features = corr_target[corr_target>0.1]
relevant_features
```

```python
sel_df = df3[['job','contact', 'month','Target','housing']]
```

```python
sel_df.head()
```

```python
#split data

df_train = sel_df.loc[sel_df['housing']=='yes'].head(200)
df_train_no = sel_df.loc[sel_df['housing']=='no'].head(200)
df_train = df_train.append(df_train_no).reset_index().drop('index',1)
```

```python
df_test = sel_df.loc[sel_df['housing']=='yes'].tail(40)
df_test_no = sel_df.loc[sel_df['housing']=='no'].tail(40)
df_test = df_test.append(df_test_no).reset_index().drop('index',1)
```

```python
for columnName, columnData in df_train.iteritems():
  print(f'Column name : {columnName}')
  print(f'Column unique : {set(columnData.values)} \n')
```

```python
#tentukan config 4.5
config = {'algorithm':'C4.5'}

#tentukan model c4.5 model di training set
c45_model2 = chef.fit(df_train, config, target_label='housing')
```

```python
with open('/content/outputs/rules/rules.py',encoding='utf8') as f:
  for line in f:
    print(line.strip())
```

```python
#prediction berdasar row
r = int(input('select row: '))
print(f'\nRow: \n {df_test.loc[r]}')
print(f'Actual: {df_test['housing'].loc[r]}')
print(f'\nPrediction: {chef.predict(c45_model2,df_test.loc[r])}')
```

```python
predicted = []
for r in df_test.index:
  predicted.append(chef.predict(c45_model2,df_test.loc[r]))
```

```python
#evaluate

from sklearn.metrics import *
print(classification_report(df_test['housing'].values,predicted,zero_division=0))
```
