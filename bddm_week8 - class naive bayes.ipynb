###BINARY CLASSIFICATION



```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import *
```

```python
df = pd.read_csv('/content/Social_Network_Ads.csv')
df.head()
```

```python
#X = variabel independen berisi Age dan Estimated Salary
X = df.iloc[:,[2,3]].values

#Y = vaiabel dependen berisi Purchased
y = df.iloc[:,-1].values
```

```python
print('X shape: ',X.shape)
print('y shape: ',y.shape)
```

```python
print('X: \n', X[:5])
print('y: \n', y[:5])
```

###data split

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size = 0.2, random_state = 0)
```

```python
print('X_train shape: ',X_train.shape)
print('X_test shape: ',X_test.shape)
print('y_train shape: ',y_train.shape)
print('y_test shape: ',y_test.shape)
```

```python
print('X_train : \n', X_train[:5])
print('y_train : \n', y_train[:5])
```

###classifier model

```python
#traububg the NB model on the Training set
classifier_model_binary = GaussianNB()
classifier_model_binary.fit(X_train,y_train)
```

###predict test data

```python
#predicting the test
y_pred = classifier_model_binary.predict(X_test)
```

```python
y_pred
```

```python
for i,y in enumerate(y_pred[:5]):
  print(f'index {i} ==> {y}')
```

```python
classifier_model_binary.predict_proba(X_test)
```

```python
def predict_one(age,salary):
  array_input = [[age,salary]]
  print(f'Probability : {classifier_model_binary.predict_proba(array_input)}')
  print(f'Predict : {classifier_model_binary.predict(array_input)}')
```

```python
predict_one(30,87000)
```

```python
#age = 18, salary = 210000
predict_one(18,210000)
```

###evaluate

```python
print(classification_report(y_test,y_pred,zero_division=0))
```

```python
# making the confusion matrix
cm = confusion_matrix(y_test,y_pred)
cm
```

```python
#load dataset
df_train = pd.read_excel('/content/data_train.xlsx')
df_test = pd.read_csv('/content/data_test.csv')
```

```python
df_train.head()
```

```python
df_train.shape
```

```python
df_test.shape
```

```python
#split
X_train=df_train.drop(['Jurusan'],axis=1)
y_train = df_train['Jurusan']

X_test=df_test.drop(['Jurusan'],axis=1)
y_test = df_test['Jurusan']
```

```python
X_train.head()
```

```python
y_train.head()
```

```python
#create classifier model

classifier_model_multiclass = GaussianNB()
classifier_model_multiclass.fit(X_train,y_train)
```

```python
#predict data
y_pred = classifier_model_multiclass.predict(X_test)
```

```python
y_prob = classifier_model_multiclass.predict_proba(X_test)
```

```python
y_prob
```

```python
for i,y in enumerate(y_pred):
  print(f'prediction for index {i} --> {y}')
```

```python
print(classification_report(y_test,y_pred,zero_division=0))
```

```python
cm2 = confusion_matrix(y_test,y_pred)
cm2
```

```python
#predict multi class
#x = math, y = eng, z = id, a = prestasi siswa, b = prestasi sekolah
def predict_multi(x,y,z,a,b):
  array_input = [[x,y,z,a,b]]
  print(f'Probability : {classifier_model_multiclass.predict_proba(array_input)}')
  print(f'Predict : {classifier_model_multiclass.predict(array_input)}')
```
