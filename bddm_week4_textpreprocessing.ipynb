#data cleaning

```python
#cleaning non alpha-numeric
import re #import module regular expression
txt = "Hi! @Mukidi, apa kabar? #sapa_Pagi."
print(re.sub(r'[^\w]', ' ', txt))

#exclude titik dan koma
print(re.sub(r'[^.,a-zA-Z0-9 \n\.]','',txt))
```

```python
#cleaning hastags
tweet = "#AndaiSajaIaTahu #ApaYangAkuRasah... #AlayersTweet #d2d"

getHashtag = re.compile(r"#(\w+)")
print("Tags = {0}".format(re.findall(getHashtag,tweet)))
```

```python
#extracing / replacing email
emailPattern = re.compile(r'[\w._%+-]+@[\w\.-]+\.[a-zA-Z]+')

txt = 'contact kami di admin@nlpindonesia.org, nlp.indonesia@sci.yahoo.com atau nlp_nusantara@internet.net'
print(re.sub(emailPattern, ' ', txt))
emailS=re.findall(emailPattern,txt)
print('email ditemukan: ', str(emailS))
```

```python
kalimat = 'Terdapat 5 jenis 6 8 ikan di kolam itu, lele, gurami, nila, patin, mas'
print(re.sub(r'\d','',kalimat))
#print(re.sub(r'\D','',kalimat))
print(re.findall(r'\d',kalimat))
```

```python
#url
urlPattern = re.compile(r'http[s]?://(?:[a-zA-Z]|[0-9]|[$-_@.&+]|[!*\(\),]|(?:%[0-9a-fA-F][0-9a-fA-F]))+')

txt = 'kunjungi web saya di https://cv.azizfath.my.id'

print("clean string:",re.sub(urlPattern,'',txt))
urls= re.findall(urlPattern,txt)
print("url yang ditemukan",urls)
```

#tokenizing

tokenizing adl proses pemisahan teks menjadi potongan-potongan yang disebut sebagai token untuk kemudian di analisa. kata,angka,simbol tanda baca dan entitas penting lainnya dapat dianggap sebagai token. didalam nlp, token diartikan sebagai "kata" meskipun tokenize juga dapat dilakukan pada paragraf maupun kalimat

```python
!pip install nltk
```

```python
import nltk
nltk.download('punkt')
nltk.download('stopwords')
```

```python
from nltk.tokenize import word_tokenize
kalimat = "Kita sedang belajar Data Mining, materi pre-processing"

tokens = nltk.tokenize.word_tokenize(kalimat)
print(tokens)
```

```python
from nltk.probability import FreqDist
kalimat = '''Lala sering sekali berbelanja online terutama saat pandemi.
Menurut lala belanja online lebih praktis daripada belanja di pasar'''

tokens = nltk.tokenize.word_tokenize(kalimat.lower())
kemunculan = nltk.FreqDist(tokens)
print(kemunculan.most_common())
```

```python
from nltk.tokenize import sent_tokenize
kalimat = '''Kita sedang belajar Data Mining, materi pre-processing.
sub materi tokenizing. menyenangkan bukan'''
tokens=nltk.tokenize.sent_tokenize(kalimat)
print(tokens)
```

#filtering

```python
from nltk.corpus import stopwords
kalimat = '''Lala sering sekali berbelanja online terutama saat pandemi.
Menurut lala belanja online lebih praktis daripada belanja di pasar'''

tokens = word_tokenize(kalimat)
listStopword = set(stopwords.words('indonesian'))

removed=[]
for t in tokens:
  if t not in listStopword:
    removed.append(t)
print(removed)
```
